# Delete Outliers

```python
def remove_outliers(df,columns,n_std):
    for col in columns:
        print('Working on column: {}'.format(col))
        
        mean = df[col].mean()
        sd = df[col].std()
        
        df = df[(df[col] <= mean+(n_std*sd))]
        
    return df
```

```python
numrical_features = ["number of adults","number of children", "number of weekend nights",
                     "number of week nights", "car parking space", "lead time","repeated",
                     "P-C","P-not-C", "average price ","special requests"]
# hotel_data = remove_outliers(hotel_data,numrical_features,3)
```

# Get output label

```python
hotel_data["booking status"] = hotel_data["booking status"].replace("Not_Canceled", 0)
hotel_data["booking status"] = hotel_data["booking status"].replace("Canceled", 1)
y_output = hotel_data.pop("booking status")
```

```python
y_output = np.array(y_output).reshape(-1, 1)
y_output
```

# Get features

```python
catgorical_features = ["type of meal","room type","market segment type"]

features = []
for cat in catgorical_features:
    vocab = hotel_data[cat].unique().reshape(-1,1)
    enc = OrdinalEncoder()
    enc.fit(vocab)
    features.append(enc.transform(np.array(hotel_data[cat]).reshape(-1,1)))

for num in numrical_features:
    features.append(np.array(hotel_data[num]).reshape(-1,1))
```

```python
features = np.concatenate(features, axis=1)
```

# data_Normlization

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit(features)
features = scaler.transform(features)
```

# feature selection

```python
from sklearn.feature_selection import chi2

chi_score = chi2(features, y_output)
p_value = pd.Series(chi_score[1], index= catgorical_features + numrical_features)
p_value.sort_values(ascending=False, inplace=True)
p_value.plot.bar()
```

```python
new_feature = np.delete(features, [4], axis=1) # delete num of childerns
```

# Data Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(new_feature, y_output, test_size=0.33, random_state=42)
```

# Gridseach_logreg

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
grid = [{'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']},
              {'penalty':['none', 'elasticnet', 'l1', 'l2']},
              {'C':np.logspace(0, 100, 10)}]

logreg = LogisticRegression()
logreg_cv=GridSearchCV(logreg,grid,cv=10, scoring='accuracy', verbose=0)
logreg_cv.fit(X_train,y_train)

```

```python
print(logreg_cv.best_params_)
print(logreg_cv.best_score_)
```

```python
from sklearn.metrics import recall_score, precision_score

clf = LogisticRegression(C=10).fit(X_train, y_train)
score = clf.score(X_test,y_test) 
print(score)

# confusion matrix
pred = clf.predict(X_test) 
clf_recall = recall_score(y_test, pred)
print(clf_recall)

clf_precision = precision_score(y_test, pred)
print(clf_precision)
```
